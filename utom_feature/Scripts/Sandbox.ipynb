```python
frontend_task_templates = [
    {
        "task_name": "Frontend implementation task name",
        "description": "Detailed description of the frontend implementation task",
        "technical_requirements": {
            "components": ["List of React components needed"],
            "libraries": ["Required frontend libraries/frameworks"],
            "api_integrations": ["Required API endpoints to integrate with"]
        },
        "execution_steps": [
            "Step-by-step implementation guide",
            "Component architecture details",
            "State management approach"
        ],
        "acceptance_criteria": [
            "UI/UX requirements that must be met",
            "Performance criteria",
            "Cross-browser compatibility requirements",
            "Responsive design requirements"
        ],
        "testing_requirements": [
            "Unit test scenarios",
            "Integration test requirements",
            "E2E test cases"
        ]
    }
]

def generate_frontend_tasks_from_context(user_story_feature_context_dict, frontend_task_templates, user_story, openai_api_key):
    """
    Generates frontend engineering tasks based on feature context and user story using OpenAI API
    
    Args:
        user_story_feature_context_dict (dict): Technical context and feature details
        frontend_task_templates (dict): Template for frontend task structure 
        user_story (dict): User story details
        openai_api_key (str): OpenAI API key
        
    Returns:
        list: List of frontend task dictionaries
    """
    client = OpenAI(api_key=openai_api_key)
    
    prompt = f"""Given the following context:
    
    Feature Context: {user_story_feature_context_dict}
    User Story: {user_story}
    
    Act as a senior frontend engineer and create detailed implementation tasks that will guide the frontend engineers in building the user interface and interactions for this user story. Focus on:
    1. Component architecture and reusability
    2. State management and data flow
    3. API integrations
    4. Performance optimization
    5. Responsive design
    6. Testing strategy
    7. Accessibility requirements
    
    Generate a list of frontend engineering tasks based on this context. The tasks should follow the template structure below:
    {frontend_task_templates}

    Output a list of json objects with each json object representing a task. Only output the list of json objects, nothing else.
    Ensure tasks are broken down into manageable chunks that can be completed in 1-2 days each.
    """

    response = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],
        max_tokens=10000,
    )

    content = response.choices[0].message.content.strip()
    
    try:
        parsed_content = json.loads(content)
        return parsed_content.get("tasks", [])
    except Exception as e:
        print(f"Error parsing response: {e}")
        try:
            parsed_content = parse_tasks_string(content)
            return parsed_content
        except Exception as e:
            print(f"Error parsing response: {e}")
            return []

def parse_tasks_string(content):
    """
    Parse a string containing a list of JSON objects and return the list.
    
    Args:
        content (str): String containing JSON array of task objects
        
    Returns:
        list: List of parsed JSON task objects, or empty list if parsing fails
    """
    try:
        # Try to parse the content directly as a JSON array
        tasks = json.loads(content)
        if isinstance(tasks, list):
            return tasks
        return []
    except json.JSONDecodeError:
        try:
            # Try to extract JSON array if embedded in other text
            start = content.find('[')
            end = content.rfind(']')
            
            if start != -1 and end != -1:
                json_str = content[start:end+1]
                tasks = json.loads(json_str)
                if isinstance(tasks, list):
                    return tasks
            return []
        except Exception as e:
            print(f"Error parsing tasks string: {e}")
            return []
```

```python
"""
MongoDB Helper Functions
"""
def get_user_stories_by_feature(client, feature_id):
    db = client['utom_features']
    return list(db.user_stories.find({'feature_id': feature_id}))

def get_features_by_project(client, project_id):
    db = client['utom_features']
    return list(db.project_features.find({'project_id': project_id}))

def get_projects_by_workspace(client, workspace_id):
    db = client['utom_features']
    return list(db.project_features.distinct('project_metadata', {'workspace_id': workspace_id}))

def update_user_story(client, story_id, updated_data):
    db = client['utom_features']
    db.user_stories.update_one({'_id': story_id}, {'$set': updated_data})

def update_feature(client, feature_id, updated_data):
    db = client['utom_features']
    db.project_features.update_one({'_id': feature_id}, {'$set': updated_data})

def update_project(client, project_id, updated_data):
    db = client['utom_features']
    db.project_features.update_many(
        {'project_id': project_id}, 
        {'$set': {'project_metadata': updated_data}}
    )

def update_workspace(client, workspace_id, updated_data):
    db = client['utom_features']
    db.project_features.update_many(
        {'workspace_id': workspace_id},
        {'$set': {'workspace_metadata': updated_data}}
    )

def delete_user_story(client, story_id):
    db = client['utom_features']
    db.user_stories.delete_one({'_id': story_id})

def delete_feature(client, feature_id):
    db = client['utom_features']
    db.project_features.delete_one({'_id': feature_id})
    db.user_stories.delete_many({'feature_id': feature_id})

def delete_project(client, project_id):
    db = client['utom_features']
    features = get_features_by_project(client, project_id)
    for feature in features:
        delete_feature(client, feature['_id'])
    db.project_features.delete_many({'project_id': project_id})

def delete_workspace(client, workspace_id):
    db = client['utom_features']
    db.project_features.delete_many({'workspace_id': workspace_id})
    db.user_stories.delete_many({'workspace_id': workspace_id})
```

```python
from playwright.async_api import async_playwright
import asyncio
import time
import json
from typing import List, Dict

class GoogleMapsScraper:
    def __init__(self):
        """Initialize basic attributes."""
        self.playwright = None
        self.browser = None
        self.context = None
        self.page = None

    async def initialize(self):
        """Async initialization of Playwright components."""
        self.playwright = await async_playwright().start()
        self.browser = await self.playwright.chromium.launch(headless=False)
        self.context = await self.browser.new_context()
        self.page = await self.context.new_page()
        return self

    async def search_places(self, query: str, max_results: int = 5) -> List[Dict]:
        """Search for places on Google Maps and extract details."""
        try:
            # Navigate to Google Maps
            await self.page.goto('https://www.google.com/maps')
            
            # Accept cookies if the dialog appears
            try:
                await self.page.click('button:has-text("Accept all")', timeout=5000)
            except:
                pass

            # Find and fill the search box
            await self.page.fill('input#searchboxinput', query)
            await self.page.keyboard.press('Enter')
            
            # Wait for results to load
            await asyncio.sleep(3)
            
            places = []
            processed_places = 0
            
            while processed_places < max_results:
                # Get all place cards
                cards = await self.page.query_selector_all('div[role="article"]')
                
                if not cards:
                    break
                
                for card in cards:
                    if processed_places >= max_results:
                        break
                        
                    try:
                        # Click on the card to open details
                        await card.click()
                        await asyncio.sleep(2)
                        
                        # Extract place details
                        place = await self._extract_place_details()
                        if place:
                            places.append(place)
                            processed_places += 1
                        
                        # Go back to results
                        await self.page.click('button[jsaction="pane.back"]')
                        await asyncio.sleep(1)
                        
                    except Exception as e:
                        print(f"Error processing place: {str(e)}")
                        continue
                
                # Scroll to load more results if needed
                if processed_places < max_results:
                    await self._scroll_results()
                    await asyncio.sleep(2)
            
            return places
            
        except Exception as e:
            print(f"Error during search: {str(e)}")
            return []

    async def _extract_place_details(self) -> Dict:
        """Extract details from the currently opened place."""
        try:
            # Extract basic information
            name = await self._safe_extract('h1.DUwDvf')
            address = await self._safe_extract('button[data-item-id="address"]')
            rating = await self._safe_extract('div.F7nice span.ceNzKf')
            
            # Extract photos
            photos = []
            try:
                # Click on the main photo to open gallery
                await self.page.click('button[jsaction="pane.heroHeaderImage.click"]')
                await asyncio.sleep(2)
                
                # Get photo URLs
                photo_elements = await self.page.query_selector_all('div[role="img"]')
                for photo in photo_elements[:5]:
                    style = await photo.get_attribute('style')
                    if "url" in style:
                        url = style.split('url("')[1].split('")')[0]
                        photos.append(url)
                
                # Close gallery
                await self.page.keyboard.press('Escape')
                await asyncio.sleep(1)
                
            except Exception as e:
                print(f"Error extracting photos: {str(e)}")
            
            # Extract reviews
            reviews = []
            try:
                # Click to expand reviews
                await self.page.click('button[jsaction="pane.rating.moreReviews"]')
                await asyncio.sleep(2)
                
                # Get review elements
                review_elements = await self.page.query_selector_all('div.jJc9Ad')
                
                for review in review_elements[:5]:
                    author = await (await review.query_selector('div.d4r55')).inner_text()
                    stars = len(await review.query_selector_all('img[src*="star_"]'))
                    text = await (await review.query_selector('span.wiI7pd')).inner_text()
                    
                    reviews.append({
                        'author': author,
                        'rating': stars,
                        'text': text
                    })
                
                # Close reviews
                await self.page.keyboard.press('Escape')
                
            except Exception as e:
                print(f"Error extracting reviews: {str(e)}")
            
            return {
                'name': name,
                'address': address,
                'rating': rating,
                'photos': photos,
                'reviews': reviews
            }
            
        except Exception as e:
            print(f"Error extracting place details: {str(e)}")
            return {}

    async def _safe_extract(self, selector: str) -> str:
        """Safely extract text from an element."""
        try:
            element = await self.page.query_selector(selector)
            return await element.inner_text() if element else ""
        except:
            return ""

    async def _scroll_results(self):
        """Scroll the results panel to load more places."""
        await self.page.evaluate('''
            document.querySelector('div[role="feed"]').scrollTop += 500
        ''')

    async def close(self):
        """Close browser and playwright instance."""
        if self.context:
            await self.context.close()
        if self.browser:
            await self.browser.close()
        if self.playwright:
            await self.playwright.stop()

async def main():
    # Initialize scraper
    scraper = GoogleMapsScraper()
    await scraper.initialize()
    
    try:
        # Example search
        results = await scraper.search_places("restaurants in New York", max_results=5)
        
        # Save results to file
        with open('places.json', 'w', encoding='utf-8') as f:
            json.dump(results, f, ensure_ascii=False, indent=2)
        
        # Print results
        for place in results:
            print("\n" + "="*50)
            print(f"Name: {place['name']}")
            print(f"Address: {place['address']}")
            print(f"Rating: {place['rating']}")
            
            print("\nPhotos:")
            for photo_url in place['photos']:
                print(f"- {photo_url}")
                
            print("\nReviews:")
            for review in place['reviews']:
                print(f"- {review['author']} ({review['rating']} stars): {review['text'][:100]}...")
    
    finally:
        await scraper.close()

# Run in Jupyter notebook
await main()
```

```python
from playwright.async_api import async_playwright
import asyncio
import json

class GoogleMapsScraper:
    def __init__(self):
        self.playwright = None
        self.browser = None
        self.context = None
        self.page = None

    async def initialize(self):
        self.playwright = await async_playwright().start()
        self.browser = await self.playwright.chromium.launch(headless=False)
        self.context = await self.browser.new_context()
        self.page = await self.context.new_page()
        return self

    async def search_places(self, query: str):
        try:
            # Navigate to Google Maps
            await self.page.goto('https://www.google.com/maps')
            print("1. Navigated to Google Maps")
            
            # Wait for search box and fill it
            await self.page.wait_for_selector('input#searchboxinput')
            await self.page.fill('input#searchboxinput', query)
            await self.page.keyboard.press('Enter')
            print(f"2. Searched for: {query}")
            
            # Wait for results to load
            print("3. Waiting for results...")
            await self.page.wait_for_selector('div[role="article"]', timeout=10000)
            print("4. Found first results")
            
            all_places = []
            
            # Scroll 3 times
            results_panel = await self.page.wait_for_selector('div[role="feed"]')
            if results_panel:
                for i in range(3):
                    # Wait a bit for content to load
                    await asyncio.sleep(2)
                    
                    # Get the visible content
                    places = await self.page.evaluate('''
                        () => {
                            const places = [];
                            const items = document.querySelectorAll('div[role="article"]');
                            
                            items.forEach(item => {
                                try {
                                    // Log the item's HTML for debugging
                                    console.log('Processing item:', item.innerHTML);
                                    
                                    const place = {
                                        name: item.querySelector('[role="heading"]')?.innerText || null,
                                        rating: item.querySelector('span[aria-label*="rating"]')?.innerText || null,
                                        reviews: item.querySelector('span[aria-label*="reviews"]')?.innerText || null,
                                        details: Array.from(item.querySelectorAll('div[role="link"] div:not([role="heading"])')).map(el => el.innerText).filter(text => text)
                                    };
                                    
                                    if (place.name) {
                                        places.push(place);
                                    }
                                } catch (e) {
                                    console.error('Error parsing place:', e);
                                }
                            });
                            
                            return places;
                        }
                    ''')
                    
                    print(f"Found {len(places)} places in scroll {i+1}")
                    
                    # Add new places to our list
                    for place in places:
                        if place not in all_places:  # Avoid duplicates
                            all_places.append(place)
                    
                    # Scroll the panel
                    await results_panel.evaluate('el => el.scrollTo(0, el.scrollHeight)')
                    print(f"Scrolled {i+1} times")
                    await asyncio.sleep(2)  # Wait for new content to load
            
            return all_places
            
        except Exception as e:
            print(f"Error during search: {str(e)}")
            return []

    async def close(self):
        if self.context:
            await self.context.close()
        if self.browser:
            await self.browser.close()
        if self.playwright:
            await self.playwright.stop()

async def main():
    scraper = GoogleMapsScraper()
    await scraper.initialize()
    
    try:
        places = await scraper.search_places("restaurants in New York")
        
        # Print results
        print("\nFound places:")
        for i, place in enumerate(places, 1):
            print(f"\n{i}. {place['name']}")
            if place['rating']:
                print(f"   Rating: {place['rating']}")
            if place['reviews']:
                print(f"   Reviews: {place['reviews']}")
            if place['details']:
                print(f"   Details: {' | '.join(place['details'])}")
        
        # Save to JSON
        with open('places.json', 'w', encoding='utf-8') as f:
            json.dump(places, f, ensure_ascii=False, indent=2)
            print(f"\nSaved {len(places)} places to places.json")
        
    finally:
        await scraper.close()

# Run in Jupyter notebook
await main()
```

```python
from playwright.async_api import async_playwright
import asyncio

async def scrape_google_maps_reviews(location: str):
    async with async_playwright() as p:
        try:
            # Launch browser in visible mode
            browser = await p.chromium.launch(headless=False)
            page = await browser.new_page()

            # Navigate to Google Maps
            await page.goto('https://www.google.com/maps')
            print("1. Navigated to Google Maps")

            # Wait for and fill search box
            await page.fill('input#searchboxinput', location)
            await page.keyboard.press('Enter')
            print(f"2. Searching for: {location}")

            # Wait for search results and click first result
            await page.wait_for_selector('div[role="article"]')
            await page.click('div[role="article"]')
            print("3. Clicked on location")

            # Wait for and click on reviews button
            await page.wait_for_selector('button[data-tab-index="1"]')
            await page.click('button[data-tab-index="1"]')
            print("4. Clicked on reviews tab")

            # Wait for reviews to load
            await page.wait_for_selector('div.jftiEf')
            print("5. Reviews loaded")

            # Extract first 10 reviews
            reviews = await page.evaluate('''
                () => {
                    const reviewElements = document.querySelectorAll('div.jftiEf');
                    const reviews = [];
                    
                    for (let i = 0; i < Math.min(10, reviewElements.length); i++) {
                        const review = {
                            author: reviewElements[i].querySelector('.d4r55')?.textContent || 'Unknown',
                            rating: reviewElements[i].querySelector('span.kvMYJc')?.getAttribute('aria-label') || 'No rating',
                            text: reviewElements[i].querySelector('.wiI7pd')?.textContent || 'No text'
                        };
                        reviews.push(review);
                    }
                    
                    return reviews;
                }
            ''')

            print("\nReviews found:")
            for i, review in enumerate(reviews, 1):
                print(f"\nReview #{i}")
                print(f"Author: {review['author']}")
                print(f"Rating: {review['rating']}")
                print(f"Text: {review['text']}")

            # Wait additional time to see results
            await asyncio.sleep(10)
            print("\nSearch complete")

        except Exception as e:
            print(f"An error occurred: {str(e)}")
        
        finally:
            await browser.close()

async def main():
    await scrape_google_maps_reviews("Shangri La London Bridge")

# Run in Jupyter notebook
await main()
```
